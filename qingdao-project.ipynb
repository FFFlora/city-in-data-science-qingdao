```python
housedata = pd.read_csv('data/lianjia100.csv',index_col = False ,header = None)
housedata.head()
```

```python
# data cleaning 
housedata = housedata.drop(0,axis = 1)
housedata.head()
```

```python
foo = lambda x: pd.Series([i for i in x.split(' | ')])
bar = lambda x: pd.Series([i for i in x.split(' / ')])
baz = lambda x: pd.Series([i for i in x.split(' - ')])
basic = housedata[4].apply(foo)
follow = housedata[5].fillna('').apply(bar)
floor = housedata[6].fillna('').apply(baz)

#housedata.columns['小区'] = housedata[housedata.columns[0]].str.split(' ',expand=True)
#rev[rev[6].notnull()]
```

```python
def shift_row(df,index,content):
    for i in range(1,6):
        df[i][index] = df[i+1][index]
        if (i==5):
            df[i+1][index]=content
    return df
shift_row(basic,956,'叠拼别墅')
shift_row(basic,1281,'叠拼别墅')
shift_row(basic,2700,'联排别墅')
```
