```python
import pandas as pd 
import pandas_profiling as ppf
from pyecharts.charts import Geo
```

```python
movein = pd.read_csv('data/rawmove_in.csv',index_col=False)
movein
```

```python
movein = pd.read_csv('data/rawmove_in.csv',index_col=False)[['city_name','value']]
moveout = pd.read_csv('data/rawmove_out.csv',index_col=False)[['city_name','value']]
```

```python
movein = movein[movein['city_name']!='city_name']
moveout = moveout[moveout['city_name']!='city_name']
moveout
```

```python
movein.value = movein.value.astype(float)
moveout.value = moveout.value.astype(float)
```

```python
top_in = pd.DataFrame(movein.groupby('city_name')['value'].mean().round(2).sort_values(ascending=False))
top_out = pd.DataFrame(moveout.groupby('city_name')['value'].mean().round(2).sort_values(ascending=False))
top_in
```

```python
# get the city pairs for migration in and out
qingdao = ['青岛','青岛','青岛','青岛','青岛','青岛','青岛','青岛','青岛','青岛']
city_pair_in = zip(qingdao,top_in.index)
city_pair_out = zip(top_out.index,qingdao)
city_pair_in= list(city_pair_in)
city_pair_out= list(city_pair_out)
city_pair_out
```

```python
# get migration cities with value 
movein_data = zip(top_in.index,top_in.value)
moveout_data = zip(top_out.index,top_out.value)
movein_data = list(movein_data)
moveout_data = list(moveout_data)
movein_data
```

```python
from example.commons import Collector, Faker
from pyecharts import options as opts
from pyecharts.charts import Geo, Page
from pyecharts.globals import ChartType, SymbolType
```

```python
c = (
    Geo()
    .add_schema(
        maptype="china",
        itemstyle_opts=opts.ItemStyleOpts(color="#e0684a", border_color="#111"),
    )
    .add(
        "",
        movein_data,
        type_=ChartType.EFFECT_SCATTER,
        color="white",label_opts=opts.LabelOpts(is_show=False)
    )
    .add(
        "Migrated In",
        city_pair_in,
        type_=ChartType.LINES,
        effect_opts=opts.EffectOpts(
            symbol=SymbolType.DIAMOND, symbol_size=6, color="yellow"
        ),
        linestyle_opts=opts.LineStyleOpts(curve=0.3),label_opts=opts.LabelOpts(is_show=True)
    )    
    .add(
        "Migrated Out",
        city_pair_out,
        type_=ChartType.LINES,
        effect_opts=opts.EffectOpts(
            symbol=SymbolType.DIAMOND, symbol_size=6, color="yellow"
        ),
        linestyle_opts=opts.LineStyleOpts(curve=0.3),label_opts=opts.LabelOpts(is_show=False)
    )
    
    #.set_series_opts(label_opts=opts.LabelOpts(is_show=True))
    .set_global_opts(title_opts=opts.TitleOpts(title="Qingdao Migration Trends \nTop 10 Cities \nfrom 08/08/2018 to 08/08/2019"),
                     legend_opts=opts.LegendOpts(selected_mode="single"),)
    
)
c.render()
```
